```python
import pandas as pd
import numpy as np

from sklearn.cluster import AffinityPropagation
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabaz_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import normalize

from agglomerative import get_data
from cluster_visual import plot_tsne_scatter

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Constants - The master stat (feature) list, and the corresponding player data.
stats_list = ['MP_x','FG','FGA','2P','2PA','3P','3PA','FT','FTA','ORB', 'DRB', 'TRB', 'AST', 'STL','BLK','TOV','PF','PTS','FG%','2P%','3P%', 'eFG%', 'FT%', 'TS%_x', 'PER', '3PAr','FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%','ORtg', 'DRtg', 'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM','VORP', 'dist', 'spd', 'tchs', 'pass', 'sast', 'ftast', 'dfgm','dfga']
df = pd.read_csv('../data/final_stats.csv')
df.set_index('player_year',inplace=True)
df.fillna(value=0,inplace=True)
```

```python
# After lot's of feature space randomization and analysis, let's use what we've learned to craft a feature space
# manually.  Here's our (tentative) new stats list:
new_stats_list = ['MP_x','tchs','2P','3P','3PA','FTr','TS%_x','ORB','TRB','AST%','STL%','BLK%','3PAr','DWS','DBPM','dist','dfga']
```

```python
# Let's check how well that feature space should cluster:
X = df[new_stats_list]
X.fillna(value=0,inplace=True)
X.info()
```

```python
plot_tsne_scatter(X,filename="final_feature_space",title="Final Feature Space")
```

```python
# Let's optimize an Affinity Propagation model:
X = df[new_stats_list]

pca = PCA(n_components=2,whiten=True,svd_solver='full')
reduced = pca.fit_transform(X)

damping_list = np.linspace(0.5,0.98,20)
preference_list = np.arange(-300,300,25)

for d in damping_list:
    print("damping = {}".format(d))
    for p in preference_list:
        aff = AffinityPropagation(damping=d,preference=p)
        aff.fit(reduced)
        try:
            silhouette = silhouette_score(reduced,aff.labels_)
            calinski = calinski_harabaz_score(reduced,aff.labels_)
        except:
            continue

        X['label'] = aff.labels_
        n_clusters = len(np.unique(aff.labels_))

        if X.loc['Nikola Jokic/2016-17']['label'] == X.loc['Marc Gasol/2016-17']['label']:
            print("preference: {} | silhouette: {} | calinski: {} | {} clusters".format(p,silhouette,calinski,n_clusters))
```

We got some pretty good scores in there - let's play around more with the space around damping=0.62,n_components=2, and preference=-380.  

```python
X = df[new_stats_list]

damping_list = np.linspace(0.61,0.63,20)
pca_components = [2,3]
preference_list = np.arange(-410,-350,10)

for d in damping_list:
    print("damping = {}".format(d))
    for c in pca_components:
        for p in preference_list:
            aff = AffinityPropagation(damping=d,preference=p)
            pca = PCA(n_components=c,whiten=True,svd_solver='full')
            reduced = pca.fit_transform(X)
            aff.fit(reduced)
            try:
                silhouette = silhouette_score(reduced,aff.labels_)
                calinski = calinski_harabaz_score(reduced,aff.labels_)
            except:
                continue

            X['label'] = aff.labels_

            if X.loc['Nikola Jokic/2016-17']['label'] == X.loc['Marc Gasol/2016-17']['label'] and silhouette > 0.4:
                print("{} components| preference: {} | silhouette: {} | calinski: {}".format(c,p,silhouette,calinski))
                print("Silhouette * Calinski = {}".format(silhouette*calinski))
```

Cool - let's run just one more to *really* hone in on these optimal parameters!

```python
X = df[new_stats_list]

damping_list = np.linspace(0.65,0.68,20)
pca_components = [2,3]
preference_list = np.arange(-425,-375,10)

for d in damping_list:
    print("damping = {}".format(d))
    for c in pca_components:
        for p in preference_list:
            aff = AffinityPropagation(damping=d,preference=p)
            pca = PCA(n_components=c,whiten=True,svd_solver='full')
            reduced = pca.fit_transform(X)
            aff.fit(reduced)
            try:
                silhouette = silhouette_score(reduced,aff.labels_)
                calinski = calinski_harabaz_score(reduced,aff.labels_)
            except:
                continue

            X['label'] = aff.labels_

            if X.loc['Nikola Jokic/2016-17']['label'] == X.loc['Marc Gasol/2016-17']['label']:
                print("{} components| preference: {} | silhouette: {} | calinski: {}".format(c,p,silhouette,calinski))
```

Looks like we're still getting our results around damping=0.677, n=2, and p=-390ish.  As in our feature exploration, we can expect the algorithm to over-cluster.  We can compensate for this by cranking down the preference value.
